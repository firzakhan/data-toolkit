#Q2.How does broadcasting work in NumPy?


Ans.NumPy's broadcasting mechanism enables efficient operations on arrays with differing shapes by automatically adjusting them to compatible dimensions, eliminating the need for manual resizing or loops.

Rules of broadcasting

Broadcasting follows a specific set of rules to determine compatibility between array shapes for element-wise operations. The process involves:

Rank Adjustment: Adding dimensions of size one to the left of the array with fewer dimensions until both arrays have the same number of dimensions.
Dimension Compatibility: Comparing the sizes of corresponding dimensions starting from the rightmost dimension.

Compatibility Condition: Dimensions are compatible if they are equal or if one of them is size 1.

Broadcasting Axis: A dimension of size 1 is conceptually stretched to match the corresponding dimension's size in the other array.

Output Shape: The output array's shape is determined by the maximum size of the corresponding dimensions of the input arrays.

Error Handling: A ValueError is raised if corresponding dimensions are not equal and neither is 1.

Conceptual vs. Actual Implementation

While broadcasting is often explained as "stretching" arrays, NumPy efficiently performs these operations without creating copies of data, relying on metadata and internal structures.

Examples

Examples of broadcasting include:

Adding a scalar to an array.

Broadcasting a 1D array across the rows of a 2D array.
Aligning dimensions in multi-dimensional arrays.
Broadcasting in NumPy helps streamline array operations, making code more concise and efficient for numerical computations.

#Q4.Explain the use of the groupby() method in Pandas?


Ans.The groupby() method in Pandas is a powerful and versatile tool for performing operations on subsets of your DataFrame based on shared characteristics in specified columns. It essentially divides the DataFrame into groups according to distinct values in one or more columns, then allows you to apply functions to each of these groups independently.


```python
import pandas as pd

# Sample DataFrame
data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A'],
        'Value': [10, 20, 15, 25, 30, 12]}
df = pd.DataFrame(data)

# Group by 'Category' and calculate the sum of 'Value' for each category
grouped_data = df.groupby('Category')['Value'].sum()

print(grouped_data)

# Output:
# Category
# A    37
# B    50
# C    25
# Name: Value, dtype: int64
```

In the example above, the DataFrame df is grouped by the Category column. Then, the sum() function calculates the sum of the Value column for each category.

Core principle: split-apply-combine

The groupby() method follows the "split-apply-combine" process:

Split: The DataFrame is divided into groups based on the unique values in the specified column(s). The result of this step is a DataFrameGroupBy object.

Apply: A function is applied independently to each of these smaller tables or Series representing the groups. Common functions include aggregation (e.g., sum(), mean(), count()), transformation (modifying values within each group while maintaining the original data structure), and filtering (retaining or discarding groups based on certain conditions).

Combine: The results of the applied function from each group are combined back into a new DataFrame or Series.

Key uses and advantages

Aggregation: Calculating summary statistics (mean, sum, min, max, count, etc.) for each group, which is useful for summarizing information and deriving insights from datasets.

Transformation: Performing operations on groups while maintaining the original structure of the data, such as standardizing data within groups or ranking items within groups.

Filtering: Filtering out groups based on certain conditions or retaining only the groups that meet the criteria, says InfluxData.

Efficient Data Analysis: Enables handling of large datasets efficiently by grouping data points that share common characteristics.

Time Series Analysis: Useful for exploring patterns in temporal data, such as analyzing monthly sales trends.

Examples

Group by a single column: Grouping data by a single column (e.g., grouping sales by product category) to calculate the sum of sales for each category.

Group by multiple columns: Grouping data based on multiple criteria (e.g., grouping sales by region and product type) to get more specific insights, such as total sales for each product category within each region.

Custom aggregation: Using the agg() method with dictionaries to apply different aggregation functions to different columns or to apply multiple aggregation functions to a single column.

Transformation with transform(): Applying functions to groups and returning a DataFrame with the same size and index as the original, such as normalizing data within each group.

Filtering groups with filter(): Filtering out groups that do not meet certain conditions, for instance, removing groups where the average salary is below a threshold.

The groupby() method is a cornerstone of data analysis in Pandas, allowing for insightful and efficient data manipulation and exploration.

#Q6. What are the differences between NumPy arrays and Python lists?


Ans. Differences between NumPy arrays and Python lists
Python lists and NumPy arrays are both ways to store collections of data in Python, but they have key differences in terms of data storage, homogeneity, size, performance, and functionality.

Here's a breakdown of the differences:

1. Data storage and homogeneity

Python Lists: Can store elements of different data types (e.g., integers, strings, floats) within the same list. This allows for flexibility but incurs overhead as each element needs to store additional information like its type and memory location. The elements are also not necessarily stored in contiguous memory locations.

NumPy Arrays: Require all elements to be of the same data type (e.g., all integers or all floats), leading to more efficient memory allocation. This homogeneity allows NumPy to store data in a compact, contiguous block of memory, reducing memory usage compared to Python lists.  

2. Size and flexibility

Python Lists: Are dynamic in size, meaning they can easily grow or shrink as elements are added or removed.

NumPy Arrays: Have a fixed size once created, providing better memory efficiency and performance for numerical operations.

3. Performance and efficiency

Python Lists: Tend to be slower for numerical computations, especially with large datasets, due to the overhead of dynamic typing and flexible data storage.
NumPy Arrays: Offer significantly better performance for numerical operations, particularly with large datasets. This is because NumPy is built on C, using optimized, pre-compiled code for numerical operations, and can leverage vectorized operations that apply functions to entire arrays without explicit Python loops.  

4. Functionality

Python Lists: Offer basic functionality for storing and accessing elements and a variety of built-in methods for manipulation like append, remove, and sort. They are suitable for general-purpose tasks requiring flexibility.
NumPy Arrays: Provide a rich set of specialized mathematical functions for numerical computations like linear algebra, Fourier transforms, and statistical operations. They are designed for numerical data and excel in efficiency for heavy mathematical tasks.  

5. When to use which

Python Lists: Preferred for general-purpose tasks when you need to store heterogeneous data or when the dataset is relatively small and performance is not a critical factor.

NumPy Arrays: Ideal for numerical computations, especially with large datasets, due to their optimized performance, memory efficiency, and specialized mathematical functions. They are widely used in scientific computing, data analysis, and machine learning fields.

In essence, Python lists prioritize flexibility and generality, while NumPy arrays prioritize numerical performance and memory efficiency for homogeneous data. Choosing the right data structure depends on your specific needs and priorities for the task at hand.

#Q8.What does the term “vectorized operation” mean in NumPy?


Ans. Vectorized operations in NumPy: operating on entire arrays at once
In NumPy, "vectorized operation" refers to performing operations on entire arrays (or parts of arrays) at once, rather than iterating through each element individually using traditional Python loops. This is a core concept that allows NumPy to achieve its high performance and efficiency, especially when dealing with large datasets.

How vectorized operations work

Instead of writing explicit for loops in Python, you can apply operations directly to NumPy arrays. For example, to add 5 to every element in an array, you simply write new_arr = arr + 5, rather than looping through each element and adding 5 individually. This is possible because NumPy uses highly optimized, pre-compiled code written in C and Fortran to perform these operations. When you use vectorized operations, NumPy efficiently executes these low-level routines, resulting in significantly faster computation times compared to standard Python loops.

Why vectorized operations are faster

Elimination of Python loop overhead: Python's interpreted nature and dynamic typing introduce overhead in loops. Each iteration involves checks and operations that add up. Vectorized operations bypass this by executing the loop in highly optimized C code, significantly reducing processing time.
Optimized low-level implementations: NumPy leverages libraries like BLAS and LAPACK for its core operations, which are highly optimized for numerical
computation and can take advantage of hardware-specific instructions.

Broadcasting: NumPy's broadcasting feature allows operations to be performed on arrays with different shapes and sizes, automatically aligning them and efficiently replicating data when necessary without consuming additional memory, further enhancing vectorized operations.

Benefits of vectorized operations

Faster execution: Vectorized code often executes orders of magnitude faster than code using explicit loops, especially for large datasets.

Cleaner and more readable code: Eliminating explicit loops makes the code more concise and easier to understand.

Scalability: Vectorized operations handle large datasets more efficiently than loops, making them ideal for tasks involving substantial amounts of data.

Parallelization: Vectorized operations can take advantage of parallel processing capabilities in modern CPUs, further boosting performance.

When to use vectorized operations

Vectorization should be your preferred method when working with large arrays or datasets in NumPy, especially for tasks involving element-wise operations, mathematical calculations, and data transformations. It's crucial for achieving high performance in numerical computing, data analysis, and machine learning workflows.

#Q10. What is the significance of hierarchical indexing in Pandas?


Ans. Significance of hierarchical indexing in Pandas

Hierarchical indexing, also known as MultiIndexing, is a powerful feature in the Pandas library that allows you to store and analyze data with multiple levels of indexing on one or more axes (typically rows) within a DataFrame. This becomes particularly significant when working with complex datasets that naturally exhibit a nested structure or multi-dimensional characteristics.

Here's why hierarchical indexing is so important

Representing and Analyzing Higher-Dimensional Data: MultiIndexing allows you to represent data that would otherwise require three or more dimensions within a two-dimensional DataFrame. Each level of the MultiIndex effectively adds another dimension to the data structure. Imagine data organized by country, then state, then city – a perfect scenario for a hierarchical index.
Organized and Structured Data: It helps in organizing and structuring data in a more intuitive way that reflects real-world relationships. This can be particularly beneficial for datasets with natural hierarchies like time series data (year, month, day), geographical data (country, state, city), or organizational data (department, team, individual).

Efficient Data Slicing and Selection: MultiIndexing simplifies the process of selecting subsets of your data. You can easily slice and dice data across different levels of the hierarchy using methods like .loc[] and .xs(). For instance, you could select all data for a specific year, or even a specific month across all years in a dataset indexed by 'Year' and 'Month'.

Enhanced Grouping and Aggregation: MultiIndexing greatly enhances the power and flexibility of grouping operations. You can perform aggregations (sums, averages, counts, etc.) on specific levels of the index or a combination of levels, using the .groupby() method. This allows for granular analysis while maintaining the overall hierarchical context.

Simplified Data Manipulation: Operations like reshaping data (moving index levels to columns and vice versa) become much easier with MultiIndexing, using methods like .unstack() and .stack(). This allows for viewing your data from different perspectives, which can be invaluable for gaining new insights.

Improved Readability and Understanding: By incorporating categorical information directly into the index, your code can become more concise and easier to understand, referencing specific data points using intuitive hierarchical indexing. The visual representation of multi-indexed DataFrames also aids in understanding the underlying data relationships.

Efficient Handling of Missing Data: MultiIndexing provides a structured approach to managing missing data. You can align data along multiple dimensions, making it easier to pinpoint missing values and potentially fill or interpolate them in a consistent manner, according to a LinkedIn post.
Essentially, hierarchical indexing empowers you to handle complex, multi-dimensional datasets within a familiar 2D structure (Series or DataFrame) while maintaining clarity, efficiency, and flexibility in your data analysis and manipulation workflows

#Q12.What is the purpose of the describe() function in Pandas?


Ans. The describe() function in Pandas is a fundamental tool for obtaining a quick statistical summary of your data in a DataFrame or Series. It's especially useful during the exploratory data analysis (EDA) phase to gain initial insights into the dataset's characteristics and distributions.

Here's a breakdown of its purpose:

1. Summarize numerical data

By default, describe() provides key metrics for numerical columns (e.g., integers, floats) in a DataFrame.
The default output for numerical columns includes:
count: The number of non-null (non-missing) entries in the column.
mean: The average value of the column.
std: The standard deviation, which measures the spread or dispersion of the values.
min: The minimum value in the column.
25% (Q1): The 25th percentile, indicating that 25% of the values are less than or equal to this value.
50% (Median/Q2): The 50th percentile, which is also the median of the column (half the values are below it).
75% (Q3): The 75th percentile, meaning 75% of the values are less than or equal to this value.

max: The maximum value in the column.

2. Summarize non-numerical (categorical/object) data
While describe() focuses on numerical data by default, it can also provide insights into non-numerical columns by explicitly specifying the include parameter.
For non-numeric data types (like strings or objects), the default output includes:

count: The number of non-null entries.
unique: The number of distinct (unique) values.
top: The most frequent value (mode).
freq: The frequency (count) of the most frequent value.

3. Customize the output

Percentiles: You can customize the percentiles included in the output by providing a list of desired percentiles (between 0 and 1) to the percentiles parameter. This allows you to explore the distribution tails or specific ranges of your data.

Including/Excluding Data Types: The include and exclude parameters allow you to specify which data types you want to include or exclude from the summary. For instance, df.describe(include='all') will generate statistics for all columns, regardless of their data type.

Specific Columns: You can apply describe() to a specific column or a subset of columns, allowing you to focus your analysis.

4. Aid in exploratory data analysis and data cleaning
describe() helps quickly understand the data types present, identify potential outliers (using min/max values), and assess the spread and central tendency of the data.

It can aid in detecting missing values (by comparing the count to the total number of rows), which is crucial for data cleaning and preprocessing steps.
In essence, describe() is a versatile function that provides a compact, powerful statistical summary of your DataFrame or Series, helping you rapidly grasp the fundamental characteristics of your data and prepare for further analysis and modeling.

#Q13. Why is handling missing data important in Pandas?

Here's a breakdown of why it's so important

1. Impact on model performance and accuracy

Many machine learning algorithms require complete datasets and will either fail or perform poorly if they encounter missing values during training or prediction.
Even algorithms that can handle missing values natively, like decision trees, might produce suboptimal or biased results if the missingness is not properly addressed.
Properly handling missing data, such as through imputation or appropriate deletion strategies, can lead to more accurate and robust models that generalize better to unseen data.

2. Bias prevention and preservation of data integrity

Ignoring missing data can introduce systematic bias into the analysis and results. For example, if higher-income individuals are less likely to report their income, simply deleting rows with missing income will skew the data and lead to an underestimation of average income.
Understanding the mechanisms behind missing data (MCAR, MAR, MNAR) is crucial for choosing the right strategy to minimize bias and maintain data integrity.
Careful handling of missing data helps maintain the representativeness of the dataset, ensuring that the analysis and model results accurately reflect the underlying population.
3. Maintaining statistical power and validity
Missing data can reduce the effective sample size, impacting the statistical power of tests to detect significant effects.
Imputation or thoughtful removal of missing data allows for the use of a wider range of statistical analysis techniques, preserving the dataset's sample size and improving the validity of conclusions drawn from the data.

4. Addressing challenges posed by missing data

Difficulty in analysis: Many statistical and machine learning methods require complete data, and missing values can render those methods inapplicable, according to GeeksforGeeks.
Reduced sample size: Removing rows with missing values, if done indiscriminately, can drastically reduce the dataset size, leading to a loss of valuable information and potentially impacting the reliability and accuracy of the analysis.
Distortion of data distributions: Improper imputation techniques can distort the natural distribution of the data, potentially flattening variations and impacting relationships between variables.
Increased model complexity: Handling missing data often adds steps to the preprocessing pipeline, increasing the complexity of the machine learning workflow.

5. Better decision-making

Ultimately, a clean dataset with appropriately handled missing values leads to more informed and trustworthy decisions based on accurate and reliable insights.
Ensuring data quality by addressing missing values is crucial for generating meaningful insights in various domains like statistical analysis, machine learning, business intelligence, customer analysis, and healthcare research.
In conclusion, handling missing data effectively is a cornerstone of robust data analysis and machine learning. By carefully considering the nature of the missingness, selecting appropriate techniques (like deletion, imputation, or using algorithms that handle missing data natively), and documenting the process, data scientists can ensure the integrity of their data, build accurate models, and ultimately derive more reliable insights from their datasets.

#Q15. How does NumPy handle multidimensional arrays?


Ans. How NumPy handles multidimensional arrays

NumPy provides a powerful and efficient way to work with multidimensional arrays, which are fundamental in scientific computing and data analysis. The core of NumPy's functionality lies in its ndarray (N-dimensional array) object, NumPy's array object.

Here's how NumPy handles multidimensional arrays

1. Homogeneous data types

All elements within a NumPy array must be of the same data type (e.g., integers, floats), according to the University at Buffalo.
This homogeneity allows NumPy to store data in a contiguous block of memory, leading to faster processing and more efficient use of CPU caches compared to Python lists, according to Analytics Vidhya.

2. Fixed size and shape

Once created, the total size of a NumPy array cannot be changed, according to the NumPy documentation.
The array has a shape attribute, which is a tuple describing the length of each dimension.
You can change the shape of an array using the .reshape() method or np.reshape(), but the total number of elements must remain the same, according to Hyperskill.

Pandas provides a number of functions for applying operations to data, and two commonly used ones are apply() and map(). While they can seem similar, their primary distinction lies in their scope of application and the types of operations they are best suited for.

map()

Scope: The map() function is exclusively for Pandas Series objects (which are essentially columns in a DataFrame).
Purpose: Its primary purpose is to substitute each value in a Series with another value. This substitution can be based on

A function: Each element in the Series is passed to the function, and the returned value replaces the original.

A dictionary: The dictionary acts as a lookup table, mapping existing values to new ones. Values not found in the dictionary are replaced with NaN (Not a Number).

Another Series: The index of the mapping Series should align with the values of the Series being operated on.

Element-wise: map() operates element-wise, meaning it processes each value in the Series independently of other values in the Series.

Best for: Simple, element-wise transformations on a single column, especially when mapping values from one domain to another (e.g., converting categorical values to numerical codes).

Caveat: You cannot use map() directly on an entire DataFrame.

apply()

Scope: The apply() function is more versatile and can be used on both Pandas Series and DataFrames.

Purpose: apply() is used to apply a function along an axis of a DataFrame or on the values of a Series.

Flexibility: When used on a DataFrame, it allows for operations across rows or columns, controlled by the axis parameter (0 for columns, 1 for rows).

Element-wise or Aggregating: apply() can be used for element-wise operations on a Series (similar to map()) or for more complex operations involving multiple elements at once (e.g., aggregating data across rows or columns, applying custom functions that affect multiple elements).

Best for:

Operations that require access to multiple columns or rows, or when you want to change the shape of the DataFrame.

Applying functions along an axis (rows or columns) of a DataFrame.

Complex data manipulations, such as filtering, aggregation, or transformations involving conditional logic across a DataFrame.

Performance: While apply() is more powerful, it can be slightly slower than map() for simple, element-wise operations on a Series.

In summary

map(): Use for element-wise operations on a single column (Series).
apply(): Use for more versatile operations, either element-wise on a Series or on rows/columns of a DataFrame.

The choice between map() and apply() hinges on the specific task requirements, balancing between the simplicity and speed of map() for single-column transformations versus the flexibility of apply() for more intricate operations involving rows, columns, or multiple columns together.

#Q21. Why is NumPy’s array slicing faster than Python’s list slicing?


Ans.NumPy array slicing is significantly faster than Python's built-in list slicing due to several underlying architectural and implementation differences:

1. Contiguous memory allocation

NumPy Arrays: NumPy arrays (ndarrays) store their elements in a single, contiguous (adjacent) block of memory. This means all elements are located right next to each other in memory.

Python Lists: Python lists are implemented as arrays of pointers to objects, and the objects themselves (the actual values) can be scattered in different locations in memory.

Impact on Slicing: When slicing a NumPy array, the operation essentially involves calculating memory addresses and strides (the number of bytes to jump to the next element), making it a very efficient operation. Since the data is contiguous, accessing elements sequentially is quick, as they are likely to be in the CPU's cache, according to AI Advances. Python lists, on the other hand, require dereferencing pointers for each element, leading to more memory accesses and potential cache misses, resulting in slower slicing operations.

2. Homogeneous data types

NumPy Arrays: All elements within a NumPy array must be of the same data type (e.g., all integers or all floats).
Python Lists: Python lists can store elements of different data types.
Impact on Slicing: The homogeneity of NumPy arrays allows for more compact memory storage and direct access to elements. NumPy can determine the size of each element and efficiently calculate memory addresses without needing to perform type checking for each element, according to NumPy documentation. The heterogeneous nature of Python lists requires more overhead for storing and managing different object types, slowing down slicing operations.

3. C implementation

NumPy is largely implemented in C, a compiled language, which allows for highly optimized and faster operations compared to Python, according to AltexSoft. Python lists, while also implemented in C, involve more Python-level overhead during slicing.

4. Views versus copies

NumPy slicing generally returns a view of the original array, meaning it creates a new array that points to the same data in memory as the original array, rather than creating a separate copy, according to StrataScratch. This eliminates the time-consuming process of copying data to a new memory location. In contrast, Python list slicing often creates a copy of the sliced elements, which requires allocating new memory and copying the data, making it a slower process.

In conclusion, NumPy's contiguous memory allocation, homogeneous data types, C implementation, and tendency to return views during slicing contribute to its significant speed advantage over Python list slicing, especially when working with large datasets, according to CodeChef.

Load or create your dataset: A DataFrame with numerical data is needed to calculate the correlation matrix.

Calculate the correlation matrix: Use the .corr() method on the DataFrame to compute the pairwise correlation coefficients between columns.

Create the heatmap: The seaborn.heatmap() function can be used to visualize the correlation matrix.

Customize the heatmap (optional): Adjust parameters like cmap, annot, fmt, linewidths, and linecolor to enhance the visualization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Create a sample DataFrame (replace with your actual data)
data = {
    'Feature_A': [10, 12, 15, 18, 20],
    'Feature_B': [5, 6, 8, 9, 11],
    'Feature_C': [2, 4, 3, 5, 7],
    'Feature_D': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)

# 2. Calculate the correlation matrix
correlation_matrix = df.corr()

# 3. Create a heatmap
plt.figure(figsize=(8, 6)) # Adjust figure size for better readability
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

#Q7. Write a program to perform element-wise multiplication of two NumPy arrays?


You can perform element-wise multiplication on two NumPy arrays in a couple of ways:
Using the * operator: This is the most straightforward and commonly used method. According to Design Gurus and, it multiplies each element in the first array by the corresponding element in the second array.
Using the np.multiply() function: This function achieves the same result as the * operator but is more explicit and can be useful in specific situations or for greater clarity in code.
Both approaches are illustrated in the code example below.

```python
import numpy as np

# Create two NumPy arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([10, 20, 30, 40, 50])

print("Array 1:", array1)
print("Array 2:", array2)

# Method 1: Using the * operator
result_operator = array1 * array2
print("\nElement-wise multiplication using * operator:", result_operator)

# Method 2: Using the np.multiply() function
result_function = np.multiply(array1, array2)
print("Element-wise multiplication using np.multiply() function:", result_function)
```

Important notes

Same Shape or Broadcastable Shapes: For element-wise multiplication to work, the two arrays must either have the exact same shape or compatible shapes according to NumPy's broadcasting rules and. Broadcasting allows NumPy to perform operations on arrays of different shapes by "stretching" the smaller array to match the dimensions of the larger one where applicable.
Data Types: Ensure that the arrays contain numeric data for the multiplication to function correctly.
Resultant Array: Both methods produce a new array with the results of the element-wise multiplication; they do not modify the original arrays.

This Boolean Series acts as a mask, filtering the original DataFrame to include only the rows where the mask is True.

This script effectively demonstrates filtering a Pandas DataFrame based on a column value exceeding a specified threshold using Boolean indexing.

#Q11.Perform matrix multiplication using NumPy?


Ans. Performing matrix multiplication using NumPy
You can perform matrix multiplication using NumPy in two primary ways:
Using the @ operator (Python 3.5+): This is the recommended and most Pythonic way for matrix multiplication. [3]
Using the np.dot() function: This function can be used for both dot products of 1D arrays and matrix multiplication of 2D arrays.
Using the np.matmul() function: This function is specifically designed for matrix multiplication.
All methods are illustrated in the code example below.

```python
import numpy as np

# Create two NumPy arrays (matrices)
matrix1 = np.array([
    [],
    []
])

matrix2 = np.array([
    [],
    []
])

print("Matrix 1:\n", matrix1)
print("Matrix 2:\n", matrix2)

# Method 1: Using the @ operator
result_at_operator = matrix1 @ matrix2
print("\nMatrix multiplication using @ operator:\n", result_at_operator)

# Method 2: Using the np.dot() function
result_dot_function = np.dot(matrix1, matrix2)
print("Matrix multiplication using np.dot() function:\n", result_dot_function)

# Method 3: Using the np.matmul() function
result_matmul_function = np.matmul(matrix1, matrix2)
print("Matrix multiplication using np.matmul() function:\n", result_matmul_function)
```

#Q13. Create a 3D scatter plot using Plotly?


You can create an interactive 3D scatter plot using Plotly Express, which simplifies the process significantly.

```python
import plotly.express as px
import pandas as pd
import numpy as np

# 1. Create a sample DataFrame with 3D data
np.random.seed(42) # for reproducibility
data = {
    'X_Axis': np.random.rand(100) * 10,
    'Y_Axis': np.random.rand(100) * 5,
    'Z_Axis': np.random.rand(100) * 15,
    'Size': np.random.rand(100) * 20 + 5, # for varying marker sizes
    'Color': np.random.choice(['Group A', 'Group B', 'Group C'], 100) # for coloring points
}
df = pd.DataFrame(data)

# 2. Create the 3D scatter plot using Plotly Express
fig = px.scatter_3d(df,
                    x='X_Axis',
                    y='Y_Axis',
                    z='Z_Axis',
                    color='Color',         # Color points based on 'Color' column
                    size='Size',           # Vary marker size based on 'Size' column
                    hover_name='Color',    # Display 'Color' on hover
                    title='3D Scatter Plot Example',
                    labels={'X_Axis': 'Feature X', 'Y_Axis': 'Feature Y', 'Z_Axis': 'Feature Z'},
                    opacity=0.7             # Adjust opacity of markers
                   )

# 3. Show the plot
fig.show()
```
